```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Media_Ads.csv')
df
```

Accessing labels

```python
x = np.linspace(19, 49, 400)
y = np.sin(x)
plt.plot(x, y, label='age')
plt.xlabel('Age')
plt.ylabel('Estimated salary')
plt.title('Plot of Social Media Ads')
plt.legend()
plt.show()
```

 Line markers are used to highlight specific points on a line plot.

```python
x = np.linspace(19, 49, 400)
y = np.sin(x)


plt.plot(x, y, marker='o', linestyle='-', color='b', label='Age')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot with Markers')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=8, markerfacecolor='red', markeredgewidth=2, linestyle='--', color='g', label='Age')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot with Customized Markers')
plt.legend()
plt.show()
```

Adding a legend to a plot helps identify different lines, markers, or data series within the same plot. The legend typically displays labels that correspond to different plot elements.

```python
x = np.linspace(19, 49, 400)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='age', color='blue')
plt.plot(x, y2, label='Estimated salary', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('age and estimated salary')

plt.legend()
plt.show()
```

A count plot is used to display the number of observations in each categorical bin using bars. It's particularly useful for visualizing the distribution of categorical variables.

```python
sns.countplot(x='Age', data=df)

plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Count Plot of Age')
plt.show()
```

```python
sns.countplot(y='Age', data=df)
plt.xlabel('age')
plt.ylabel('estimated salary')
plt.title('Horizontal Count Plot')
plt.show()
```
