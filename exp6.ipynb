```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_attack_dataset.csv')
df
```

**Scatter** **Plot**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Blood Pressure (mmHg)'], c='blue', marker='o')
plt.xlabel('Age')
plt.ylabel('Blood pressure')
plt.title('Scatter Plot')
plt.show()
```

**Bar** **Plot**

```python
import matplotlib.pyplot as plt
plt.bar(df['Age'],df['Blood Pressure (mmHg)'])
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Bar Plot ')
plt.show()
```

**Histogram**

A histogram is a type of bar chart that represents the distribution of numerical data. It shows how often different ranges (or "bins") of values occur in a dataset. Each bar in the histogram corresponds to a range of values, and the height of the bar indicates the frequency of data points within that range.

```python
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Histogram')
plt.show()
```

**Line** **Charts**

```python
plt.plot(df['Age'], df['Blood Pressure (mmHg)'], marker='o')
plt.xlabel('Age')
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Age Over Blood Pressure')
plt.grid(True)
plt.show()
```

**Boxplot**

```python
sns.boxplot(x='Age', y='Blood Pressure (mmHg)', data=df)
plt.xlabel('Age')
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Box Plot Example')
plt.show()
```
